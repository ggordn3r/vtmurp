```python
import numpy as np
import pandas as pd
import plotly.express as px
import datetime
```

```python
df = pd.read_excel("phillystops.xlsx")
```

```python
df.info()
```

```python
df = df.dropna(axis=0, how='any')
df['month'] = pd.DatetimeIndex(df['date']).month
df = df.drop(axis=1, labels=['date', 'location', 'latitude', 'longitude', 'servicearea'])
df.info()
```

```python
print(df.arrest.value_counts())
print(df.frisk.value_counts())
print(df.searchperson.value_counts())
```

```python
gender = pd.pivot_table(df, values=['arrest', 'frisk', 'searchperson'], index=[], columns=['gender'], aggfunc=np.sum)
gender
```

```python
# should I combine Am Indian, Unknown, and Asian as Other, or drop altogether?
# should I split this two ways, black vs. white, and latino vs. non-latino?
race = pd.pivot_table(df, values=['arrest', 'frisk', 'searchperson'], index=[], columns=['race'], aggfunc=np.sum)
race
```

```python
year = pd.pivot_table(df, values=['arrest', 'frisk', 'searchperson'], index=[], columns=['year'], aggfunc=np.sum)
year
```

```python
fig = px.line(year)
fig.show()
```

```python
cols = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
dayweek = pd.pivot_table(df, values=['arrest', 'frisk', 'searchperson'], index=[], columns=['dayweek'], aggfunc=np.sum)
dayweek = dayweek[cols]
dayweek
```

```python
fig = px.bar(dayweek)
fig.show()
```

```python
district = pd.pivot_table(df, values=['arrest', 'frisk', 'searchperson'], index=[], columns=['district'], aggfunc=np.sum)
district
```

```python
fig = px.bar(district)
fig.show()
```
